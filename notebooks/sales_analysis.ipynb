```python
import pandas as pd
```

```python
df = pd.read_csv("../data/sales_data.csv")
df
```

```python
df.groupby("region",as_index=False)["amount"].sum()
```

```python
df.groupby("month",as_index=False)["amount"].sum()
```

```python
total_sales=df["amount"].sum()
customer_sales=df.groupby("customer")["amount"].sum()
customer_sales[customer_sales>0.25*total_sales]
```

## Key Insights

- The North region generated the highest sales.
- February had higher total sales than January.
- Customer A contributed the highest share of revenue.
